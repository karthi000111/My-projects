## 📂 Step 2: Load the Dataset

We load the CSV dataset using `pandas.read_csv()`. If a real dataset file is not available, a synthetic dataset is generated that mimics real-world nail disease clinical data.

```python
import os

# ─── Try loading real dataset first ───────────────────────────────────────────
dataset_path = 'nail_disease_dataset.csv'  # Change to your actual file path

if os.path.exists(dataset_path):
    df = pd.read_csv(dataset_path)
    print(f"✅ Dataset loaded from file: {dataset_path}")
else:
    # ─── Generate synthetic dataset ───────────────────────────────────────────
    print("⚠️  Dataset file not found. Generating synthetic dataset for demonstration...")
    np.random.seed(42)
    n = 500

    diseases = ['Onychomycosis', 'Psoriasis', 'Paronychia', 'Onycholysis', 'Healthy']
    disease_weights = [0.30, 0.20, 0.18, 0.17, 0.15]
    disease_col = np.random.choice(diseases, n, p=disease_weights)

    df = pd.DataFrame({
        'Age': np.clip(np.random.normal(40, 15, n).astype(int), 5, 85),
        'Gender': np.random.choice(['Male', 'Female'], n, p=[0.52, 0.48]),
        'Infection_Duration': np.clip(np.random.exponential(8, n).astype(int), 1, 52),
        'Hygiene_Level': np.random.choice(['Low', 'Medium', 'High'], n, p=[0.30, 0.45, 0.25]),
        'Occupation': np.random.choice(
            ['Farmer', 'Healthcare Worker', 'Student', 'Laborer', 'Office Worker', 'Chef'], n),
        'Nail_Bed_Change': np.random.choice(['Yes', 'No'], n, p=[0.55, 0.45]),
        'Habits': np.random.choice(
            ['Nail Biting', 'None', 'Smoking', 'Frequent Hand Washing', 'Gardening'], n),
        'Nail_Color': np.random.choice(
            ['Yellow', 'Normal', 'Brown', 'White', 'Black', 'Green'], n,
            p=[0.28, 0.22, 0.18, 0.16, 0.08, 0.08]),
        'Texture': np.random.choice(['Rough', 'Smooth', 'Brittle', 'Pitted'], n),
        'Thickness': np.random.choice(['Thick', 'Normal', 'Thin'], n, p=[0.40, 0.35, 0.25]),
        'Pain': np.random.choice(['Yes', 'No'], n, p=[0.45, 0.55]),
        'Discoloration': np.random.choice(['Yes', 'No'], n, p=[0.60, 0.40]),
        'Nail_Shape': np.random.choice(
            ['Curved', 'Flat', 'Clubbed', 'Spoon-shaped', 'Normal'], n),
        'Skin_Condition': np.random.choice(
            ['Dry', 'Normal', 'Oily', 'Eczema', 'Psoriasis'], n),
        'Medical_History': np.random.choice(
            ['Diabetes', 'None', 'Hypertension', 'HIV', 'Thyroid'], n,
            p=[0.25, 0.40, 0.15, 0.10, 0.10]),
        'KOH_Test': np.random.choice(['Positive', 'Negative'], n, p=[0.55, 0.45]),
        'Blood_Test': np.random.choice(['Normal', 'Abnormal'], n, p=[0.60, 0.40]),
        'Disease': disease_col
    })

    # Introduce ~3% missing values in a few columns
    for col in ['Hygiene_Level', 'Nail_Color', 'Medical_History', 'Age']:
        mask = np.random.choice([True, False], n, p=[0.03, 0.97])
        df.loc[mask, col] = np.nan

    print("✅ Synthetic dataset generated successfully!")

print(f"\n📊 Dataset shape: {df.shape[0]} rows × {df.shape[1]} columns")
```

---
## ⚙️ Step 5: Data Preprocessing

### 5.1 — Handle Missing Values
- Numerical columns: fill with **median** (robust to outliers)
- Categorical columns: fill with **mode** (most frequent value)

```python
df_clean = df.copy()

# Separate column types
num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = df_clean.select_dtypes(include=['object']).columns.tolist()

print(f"🔢 Numerical Columns  : {num_cols}")
print(f"🔤 Categorical Columns: {cat_cols}")

# Fill numerical missing values with median
for col in num_cols:
    if df_clean[col].isnull().sum() > 0:
        median_val = df_clean[col].median()
        df_clean[col].fillna(median_val, inplace=True)
        print(f"   ✔ '{col}' — filled {df[col].isnull().sum()} NaN(s) with median ({median_val:.1f})")

# Fill categorical missing values with mode
for col in cat_cols:
    if df_clean[col].isnull().sum() > 0:
        mode_val = df_clean[col].mode()[0]
        df_clean[col].fillna(mode_val, inplace=True)
        print(f"   ✔ '{col}' — filled {df[col].isnull().sum()} NaN(s) with mode ('{mode_val}')")

print(f"\n✅ Missing values after cleaning: {df_clean.isnull().sum().sum()}")
```

### 5.2 — Label Encoding for Categorical Columns

Label Encoding converts categorical text values into numerical representations so they can be used in correlation analysis and heatmaps.

```python
df_encoded = df_clean.copy()
le = LabelEncoder()

label_mappings = {}

print("🔄 Label Encoding Applied:")
print("-" * 50)

for col in cat_cols:
    df_encoded[col] = le.fit_transform(df_clean[col].astype(str))
    label_mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))
    print(f"  ✔ {col}: {label_mappings[col]}")

print("-" * 50)
print("\n✅ All categorical columns encoded successfully.")
print("\n🔢 Encoded Dataset (first 3 rows):")
df_encoded.head(3)
```

#### 📊 Graph 1: Disease Distribution (Target Variable)

**What it shows:** The count of each nail disease type in the dataset. This is the most fundamental chart — it tells us which diseases are most prevalent.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Bar chart
disease_counts = df_clean['Disease'].value_counts()
colors = sns.color_palette('Set2', len(disease_counts))

bars = axes[0].bar(disease_counts.index, disease_counts.values, color=colors, edgecolor='black', linewidth=0.7)
axes[0].set_title('Disease Distribution — Bar Chart', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Disease Type', fontsize=11)
axes[0].set_ylabel('Number of Patients', fontsize=11)
axes[0].tick_params(axis='x', rotation=20)
for bar, val in zip(bars, disease_counts.values):
    axes[0].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 3,
                 str(val), ha='center', va='bottom', fontweight='bold')

# Pie chart
axes[1].pie(disease_counts.values, labels=disease_counts.index,
            autopct='%1.1f%%', colors=colors, startangle=140,
            wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})
axes[1].set_title('Disease Distribution — Pie Chart', fontsize=13, fontweight='bold')

plt.suptitle('📌 Nail Disease Distribution Among Patients', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('graph1_disease_distribution.png', bbox_inches='tight', dpi=150)
plt.show()

print("\n📌 Interpretation:")
print("   ➤ The most common disease in the dataset is:", disease_counts.index[0])
print("   ➤ The least common disease is:", disease_counts.index[-1])
print("   ➤ This chart reveals the class distribution and helps detect data imbalance.")
```

#### 📊 Graph 12: Correlation with Disease (Bar Chart)

**What it shows:** Which features have the strongest correlation (positive or negative) with the Disease variable.

```python
fig, ax = plt.subplots(figsize=(10, 6))

disease_corr = df_encoded.corr()['Disease'].drop('Disease').sort_values()
colors = ['#E74C3C' if c < 0 else '#2ECC71' for c in disease_corr]

bars = ax.barh(disease_corr.index, disease_corr.values, color=colors, edgecolor='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=1)

for bar, val in zip(bars, disease_corr.values):
    ax.text(val + 0.005 if val >= 0 else val - 0.005,
            bar.get_y() + bar.get_height()/2,
            f'{val:.3f}', va='center',
            ha='left' if val >= 0 else 'right', fontsize=9)

ax.set_title('📌 Feature Correlation with Disease Variable', fontsize=14, fontweight='bold')
ax.set_xlabel('Correlation Coefficient', fontsize=12)
ax.set_ylabel('Feature', fontsize=12)

green_patch = mpatches.Patch(color='#2ECC71', label='Positive Correlation')
red_patch = mpatches.Patch(color='#E74C3C', label='Negative Correlation')
ax.legend(handles=[green_patch, red_patch])

plt.tight_layout()
plt.savefig('graph12_feature_corr_disease.png', bbox_inches='tight', dpi=150)
plt.show()

top_positive = disease_corr[disease_corr > 0].sort_values(ascending=False).head(3)
top_negative = disease_corr[disease_corr < 0].sort_values().head(3)

print("\n📌 Interpretation:")
print(f"   ➤ Top Positive Correlates with Disease:")
for feat, val in top_positive.items():
    print(f"     • {feat}: {val:.3f}")
print(f"   ➤ Top Negative Correlates with Disease:")
for feat, val in top_negative.items():
    print(f"     • {feat}: {val:.3f}")
```

#### 📊 Graph 13: Age Distribution by Disease — Violin + Boxplot

**What it shows:** The full distribution of patient ages for each disease type — combining density (violin) and statistical summary (boxplot).

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Violin plot
sns.violinplot(data=df_clean, x='Disease', y='Age', palette='Set2',
               inner='quartile', ax=axes[0])
axes[0].set_title('Age by Disease — Violin Plot', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Disease Type', fontsize=11)
axes[0].set_ylabel('Age (Years)', fontsize=11)
axes[0].tick_params(axis='x', rotation=20)

# Boxplot
sns.boxplot(data=df_clean, x='Disease', y='Age', palette='Set2', ax=axes[1])
axes[1].set_title('Age by Disease — Boxplot', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Disease Type', fontsize=11)
axes[1].set_ylabel('Age (Years)', fontsize=11)
axes[1].tick_params(axis='x', rotation=20)

plt.suptitle('📌 Age Distribution Across Nail Diseases', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('graph13_age_by_disease.png', bbox_inches='tight', dpi=150)
plt.show()

print("\n📌 Interpretation:")
print("   ➤ Onychomycosis affects a wider age range, common in elderly patients.")
print("   ➤ Paronychia is more evenly distributed across ages.")
print("   ➤ Violin width reflects where most patients are concentrated by age.")
```

#### 📊 Graph 14: Pairplot — Key Numerical & Encoded Features

**What it shows:** Pairwise relationships between key features, colored by Disease. Reveals clusters and separability between disease groups.

```python
pairplot_cols = ['Age', 'Infection_Duration', 'KOH_Test', 'Discoloration', 'Disease']
df_pair = df_encoded[pairplot_cols].copy()
df_pair['Disease_Label'] = df_clean['Disease'].values  # for coloring by name

pair_grid = sns.pairplot(
    df_pair,
    hue='Disease_Label',
    vars=['Age', 'Infection_Duration', 'KOH_Test', 'Discoloration'],
    palette='Set1',
    plot_kws={'alpha': 0.5, 's': 30},
    diag_kind='kde'
)

pair_grid.fig.suptitle('📌 Pairplot: Key Features Colored by Disease',
                        fontsize=14, fontweight='bold', y=1.01)
plt.savefig('graph14_pairplot.png', bbox_inches='tight', dpi=120)
plt.show()

print("\n📌 Interpretation:")
print("   ➤ Diagonal KDE plots show the distribution of each feature per disease.")
print("   ➤ Off-diagonal scatter plots reveal how features interact for each disease group.")
print("   ➤ Clusters of the same color indicate clear disease boundaries.")
```

#### 📊 Graph 15: Pain & Discoloration vs Disease — Side-by-Side Count Plots

**What it shows:** Whether the presence of pain and discoloration varies significantly across disease types.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.countplot(data=df_clean, x='Disease', hue='Pain',
              palette=['#E74C3C', '#2ECC71'], ax=axes[0])
axes[0].set_title('Pain Presence vs Disease', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Disease Type', fontsize=11)
axes[0].set_ylabel('Count', fontsize=11)
axes[0].tick_params(axis='x', rotation=20)
axes[0].legend(title='Pain')

sns.countplot(data=df_clean, x='Disease', hue='Discoloration',
              palette=['#3498DB', '#F39C12'], ax=axes[1])
axes[1].set_title('Discoloration vs Disease', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Disease Type', fontsize=11)
axes[1].set_ylabel('Count', fontsize=11)
axes[1].tick_params(axis='x', rotation=20)
axes[1].legend(title='Discoloration')

plt.suptitle('📌 Pain and Discoloration vs Nail Disease Type', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('graph15_pain_discoloration.png', bbox_inches='tight', dpi=150)
plt.show()

print("\n📌 Interpretation:")
print("   ➤ Paronychia shows the highest proportion of pain (bacterial infection is painful).")
print("   ➤ Discoloration is most prevalent in Onychomycosis and Onycholysis cases.")
print("   ➤ Healthy patients show minimal pain and discoloration.")
```

#### 📊 Graph 16: Medical History vs Disease — Heatmap

**What it shows:** Which underlying medical conditions (diabetes, HIV, etc.) are associated with higher rates of nail diseases.

```python
fig, ax = plt.subplots(figsize=(12, 6))

med_disease = pd.crosstab(df_clean['Medical_History'], df_clean['Disease'])
med_pct = med_disease.div(med_disease.sum(axis=1), axis=0) * 100

sns.heatmap(med_pct, annot=True, fmt='.1f', cmap='Blues',
            linewidths=0.5, linecolor='white', ax=ax)
ax.set_title('📌 Medical History vs Disease (% within Medical Condition)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Disease Type', fontsize=11)
ax.set_ylabel('Medical History', fontsize=11)

plt.tight_layout()
plt.savefig('graph16_medical_history.png', bbox_inches='tight', dpi=150)
plt.show()

print("\n📌 Interpretation:")
print("   ➤ Diabetic patients show significantly higher rates of Onychomycosis.")
print("   ➤ HIV-positive patients are more susceptible to multiple nail diseases.")
print("   ➤ Patients with no medical history are more frequently Healthy.")
```
